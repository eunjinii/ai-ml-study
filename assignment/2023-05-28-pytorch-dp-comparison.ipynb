```python
import matplotlib.pyplot as plt

def getAverage(list):
  return round(sum(list) / len(list), 5)

train_acc_list_baseline = [0.35588, 0.48586, 0.55012, 0.58410, 0.61180]
train_time_list_baseline = [6.82, 16.82, 13.65, 11.30, 12.92]
train_acc_baseline_avg = getAverage(train_acc_list_baseline)
train_time_baseline_avg = getAverage(train_time_list_baseline)

train_acc_list_DP = [0.36860, 0.51156, 0.57416, 0.61204, 0.63652]
train_time_list_DP = [7.17, 7.37, 6.00, 6.42, 5.77]
train_acc_DP_avg = getAverage(train_acc_list_DP)
train_time_DP_avg = getAverage(train_time_list_DP)

print(f'Training Accuracy Avg(Baseline): {train_acc_baseline_avg}')
print(f'Training Accuracy Avg(DP): {train_acc_DP_avg}')
print(f'Training Time Avg(Baseline): {train_time_baseline_avg}(sec / epoch)')
print(f'Training Time Avg(DP): {train_time_DP_avg}(sec / epoch)')

training_accuracy_diff = round(train_acc_DP_avg - train_acc_baseline_avg, 3)
training_time_diff = round(train_time_DP_avg - train_time_baseline_avg, 3)

print(f"DP를 적용한 결과 Epoch 당 Training Time은 Baseline 대비 {'+' if training_time_diff > 0 else '-'}{abs(training_time_diff)} 변동")
print(f"DP를 적용한 결과 Training Accuracy는 Baseline 대비 {'+' if training_accuracy_diff > 0 else '-'}{abs(training_accuracy_diff)} 변동")

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Plot training time on the first subplot
ax1.plot(range(5), train_time_list_baseline, label='Baseline')
ax1.plot(range(5), train_time_list_DP, label='DataParallel')
ax1.set_ylabel('Training Time (s)')
# ax1.set_ylim(0,3)
ax1.legend()
ax1.grid(axis='both', linestyle='--')

# Plot training accuracy on the second subplot
ax2.plot(range(5), train_acc_list_baseline, label='Baseline')
ax2.plot(range(5), train_acc_list_DP, label='DataParallel')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Training Accuracy')
ax2.legend()
ax2.grid(axis='both', linestyle='--')

# Set overall title
fig.suptitle('Baseline vs. DataParallel Training')

# Show the plot
plt.show()
```
